```python
queryIngs = ['onion', 'hoisin sauce', 'butter lettuce', 'orange']
```

```python
def OR(list1,list2):
  i = 0
  j = 0
  answer = []
#   global computations
  while(i<len(list1) and j<len(list2)):
    if(list1[i]<list2[j]):
      answer.append(list1[i])
      i+=1
    #   computations+=1
    elif(list1[i]>list2[j]):
      answer.append(list2[j])
      j+=1
    #   computations+=2
    else:
      answer.append(list1[i])
      i+=1
      j+=1
    #   computations+=2
  while(i<len(list1)):
    answer.append(list1[i])
    i+=1
  while(j<len(list2)):
    answer.append(list2[j])
    j+=1
  return answer
```

```python
ans = []
for i in queryIngs:
    ans = OR(invertedIndex[i][1], ans)
```

```python
finalAns = []
for i in ans:
    ings = newDf.loc[i]['ingredients']
    ings = ings.split(', ')
    query = set(queryIngs)
    ings = set(ings)
    finalAns.append((i, len(query.intersection(ings))))
```

```python
def Sort_Tuple(tup):
    tup.sort(key = lambda x: x[1])
    return tup
```
